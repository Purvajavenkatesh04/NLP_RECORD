```python
import nltk
from nltk.util import ngrams
from nltk.tokenize import word_tokenize

# Ensure necessary NLTK resources are downloaded
nltk.download('punkt')

def generate_ngrams():
    # Step 1: Get input from the user
    sentence = input("Enter a sentence: ")
    n = int(input("Enter the n-gram size (e.g., 1 for unigrams, 2 for bigrams, 3 for trigrams): "))

    # Step 2: Tokenize the sentence into words
    tokens = word_tokenize(sentence)

    # Step 3: Generate n-grams
    n_grams = ngrams(tokens, n)

    # Step 4: Display the n-grams as space-separated words
    print(f"\n{n}-grams as space-separated words:")
    for ngram in n_grams:
        print(' '.join(ngram))  # Print each n-gram as a space-separated string

# Run the function
generate_ngrams()
```
