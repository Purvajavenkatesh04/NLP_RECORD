```python
import nltk
from nltk.util import ngrams
from nltk.tokenize import word_tokenize
from collections import Counter

nltk.download('punkt')

def laplace_smoothing():
    # Get user input
    sentence = input("Enter a sentence: ")
    n = int(input("Enter n-gram size: "))

    tokens = word_tokenize(sentence)  # Tokenize the sentence
    n_grams = list(ngrams(tokens, n))  # Generate n-grams
    n_gram_counts = Counter(n_grams)  # Count n-gram frequencies
    n_minus_1_gram_counts = Counter(ngrams(tokens, n-1)) if n > 1 else Counter()

    V = len(set(tokens))  # Vocabulary size
    smoothed_ngrams = [
        ' '.join(ngram)  # Join n-grams as space-separated words
        for ngram, count in n_gram_counts.items()
        if (count + 1) / (n_minus_1_gram_counts[ngram[:-1]] + V) > 0  # Only include non-zero smoothed n-grams
    ]

    # Print smoothed n-grams as words
    print("\nSmoothed n-grams as words:")
    for ngram in smoothed_ngrams:
        print(ngram)

# Run the function
laplace_smoothing()
```
