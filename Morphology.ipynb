```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

# Download required resources (only needed once)
nltk.download('punkt_tab')

# Input text
text = "The boys are playing football happily"

# Tokenization
tokens = word_tokenize(text)

# Initialize stemmer (a basic morphology tool)
stemmer = PorterStemmer()

# Apply stemming (a type of morphological analysis)
morph_tokens = [stemmer.stem(word) for word in tokens]

# Output
print("Tokenized and Morphologically processed words:")
print([token for token in morph_tokens])
```
