```python
import spacy

# Load English model
nlp = spacy.load("en_core_web_sm")

# Sample text
text = "Hello, world! This is tokenization using spaCy."

# Process the text
doc = nlp(text)

# Print each token in square brackets
print(' '.join(f'[{token.text}]' for token in doc))
```
